## Integrate summarization and text-to-speech

### Subtask:
Modify the main `summarize` function to call the text-to-speech function after generating the summary and return both the summary text and the audio file path.


**Reasoning**:
Modify the existing `summarize` function to call the `text_to_speech` function and return both the summary text and the audio file path, ensuring the return order matches the Gradio interface outputs.



```python
import gradio as gr
from transformers import pipeline
import datetime
import os
from gtts import gTTS

def text_to_speech(text):
    """Converts text to speech and saves it to a file."""
    # Generate a unique filename using a timestamp
    audio_filename = f"audio_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.mp3"

    # Instantiate gTTS with the input text and language
    tts = gTTS(text=text, lang='en')

    # Save the speech to the audio file
    tts.save(audio_filename)

    # Return the path to the saved audio file
    return audio_filename


def summarize_and_speak(input_text):
    # Load the summarization pipeline
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

    # Summarize the text
    summary = summarizer(input_text, max_length=150, min_length=100, do_sample=False)
    summary_text = summary[0]['summary_text']

    # Save summary to a temporary file
    summary_filename = "summary.txt"
    with open(summary_filename, "w") as f:
        f.write(summary_text)

    # Suffix file name with datetime stamp and rename
    summary_final_filename = "summary" + datetime.datetime.now().strftime("%Y%m%d_%H%M%S") + ".txt"
    os.rename(summary_filename, summary_final_filename)

    # Convert summary text to speech
    audio_output_path = text_to_speech(summary_text)

    # Return summary text, summary file path, and audio file path
    return summary_text, summary_final_filename, audio_output_path

demo = gr.Interface(
    fn=summarize_and_speak,
    inputs="text",
    outputs=[
        gr.Textbox(label="Summary"),
        gr.File(label="Download Summary"),
        gr.Audio(label="Generated Speech")
    ]
)

demo.launch(share=True)
```
